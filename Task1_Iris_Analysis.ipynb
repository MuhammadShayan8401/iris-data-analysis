## Introduction
The objective of this task is to explore and visualize the Iris dataset using Python.
This includes understanding the dataset, performing exploratory data analysis (EDA),
and identifying patterns using visualizations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

print(df.shape)
print(df.columns)
print(df.head())
```

The dataset contains 150 rows and 5 columns including features like sepal length,
sepal width, petal length, petal width, and species.

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df) ##scatterplot
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.hist(df['petal_length'], bins=20)  ##histogram
plt.title('Petal Length Distribution')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=df) ##boxplot
plt.title('Petal Length by Species')
plt.show()
```
